##Eliminate columns that won't be use

```python
df_public_dataset = df_public_dataset.drop(
    columns=[
        'product_name_lenght',
        'product_description_lenght',
        'product_photos_qty',
        'review_comment_title',
        'review_comment_message'
    ]
)
```

```python
df_classified = df_classified.drop(
    columns=[
        'product_name_lenght',
        'product_description_lenght',
        'product_photos_qty',
        'review_comment_title',
        'review_comment_message'
    ]
)
```

## Look for bad quality data

Check unique values vs expected


```python
print(df_public_dataset['order_status'].unique())

```

```python
print(df_classified['most_voted_class'].unique())
```

we're going to change the nan value for 'unknown' in order to specify better the situation


```python
df_classified['most_voted_class'] = df_classified['most_voted_class'].fillna('unknown')
```

we do this with subclass as well

```python
df_classified['most_voted_subclass'] = df_classified['most_voted_subclass'].fillna('unknown')
```
